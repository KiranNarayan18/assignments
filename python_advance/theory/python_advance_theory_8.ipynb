##### Q1. What are the two latest user-defined exception constraints in Python 3.X?

__cause__: This constraint is used to specify the cause of the exception, which can be another exception that caused the current one. This constraint is useful when you want to provide more context about the reason for the exception, and it can help with debugging and troubleshooting.

__context__: This constraint is used to specify the context of the exception, which can be another exception that is being handled. This constraint is useful when you want to chain exceptions together, so that the original exception is preserved and can be traced back to its source.

##### Q2. How are class-based exceptions that have been raised matched to handlers?

The interpreter looks for a try block that encloses the code that raised the exception. If a try block is found, the interpreter checks the associated except clauses to see if they match the exception type.

If the exception type matches one of the except clauses, the corresponding block of code is executed, and the exception is considered handled.

If the exception type does not match any of the except clauses in the enclosing try block, the search continues to the next enclosing try block, and so on, until the top level of the program is reached.

If no try block is found that can handle the exception, the interpreter prints an error message and terminates the program.


##### Q3. Describe two methods for attaching context information to exception artefacts.

Using the args attribute: Every exception in Python has an args attribute, which is a tuple that contains the arguments passed to the exception's constructor. You can attach context information to this tuple by adding additional items that describe the context of the error.

Creating a custom exception class: Another way to attach context information to an exception is to create a custom exception class that includes additional attributes or methods for describing the context of the error.

##### Q4. Describe two methods for specifying the text of an exception object's error message.
Passing a string to the exception constructor: The simplest way to specify the text of an exception object's error message is to pass a string to the exception constructor. The string should provide a clear and concise description of the error that caused the exception.

Subclassing an exception and overriding the __str__ method: Another way to specify the text of an exception object's error message is to define a custom exception class that subclasses the appropriate base exception class (e.g. Exception, ValueError, etc.), and override the __str__ method to return a custom error message.

##### Q5. Why do you no longer use string-based exceptions?

Lack of type checking: Since string-based exceptions don't have a specific type or class associated with them, it's difficult to perform any kind of type checking or catch specific types of exceptions. This can make it harder to handle errors in a predictable and robust way.

Lack of traceback information: When a string-based exception is raised, the traceback information that's generated by Python is less informative and harder to use for debugging purposes. This can make it harder to track down the source of an error and fix it.

Confusion with string literals: Since string-based exceptions are just string literals, it's easy to accidentally raise an exception by including a string literal in your code that's not intended to be an error. This can lead to confusion and make it harder to identify the actual errors in your code.
